```python
import numpy as np
import pandas as pd
from IPython.display import Markdown
from BitStream import BitStream
```

```python
cname1 = "currentmirror_swmxtst_schem"
cname2 = "swmxtst_npmirror"

circname = cname2
netlist_path = "netlist_catalog/" + circname + "_nl.txt"
```

```python
swmxtst = BitStream(circname)
swmxtst.netlistInput(netlist_path)

x = ""
for line in swmxtst.instance_netlist:
    s = (line[i] for i in range(len(line)))
    # print(' '.join(s))
    x += " ".join(s) + "<br>"

Markdown(
    f"""
{'*<u>stripped netlist containing only instance definition lines:</u>*'} 

{x}
"""
)
```

```python
bus_table = ""
for i in range(len(swmxtst.active_buslist)):
    pinlist = swmxtst.active_pinlists[i]
    pinlist = ", ".join(pinlist)
    bus_table += f"| {swmxtst.active_buslist[i]} | {pinlist} |\n"
Markdown(
    f"""
*<u>swmxtst:</u>*        
| bus number | active pins |
|:----------:|:-----------:|
{bus_table}
"""
)
```

```python
bitstream = swmxtst.bitstream
print(bitstream)
swmxtst.writeHexFile("bitstream_catalog/swmxtst_hxs.txt")
```

```python
import matplotlib.pyplot as plt
import libm2k
from m2k_streamout_utils import StreamOutUtils

MAX_SAMPLE_RATE = StreamOutUtils.MAX_SAMPLE_RATE
MAX_BIT_RATE = StreamOutUtils.MAX_BIT_RATE
MAX_SIM_TIME = StreamOutUtils.MAX_SIM_TIME
```

```python
dch_en_wr, dch_en_rd = (6, 5)
dch_clk_wr, dch_clk_rd = (7, 4)
dch_d_wr, dch_d_rd = (8, 3)
channels_out = [dch_en_wr, dch_clk_wr, dch_d_wr]
ctx = libm2k.m2kOpen()

if ctx is None:
    raise ConnectionError(
        "Connection Error: No ADALM2000 device available/connected to your PC."
    )

uri = ctx.getUri()
print("connected to ADALM2000 device: [" '"{}"' "]\n".format(uri))

if ctx.hasMixedSignal() is False:
    print("Error: Firmware does not support mixed signal mode.")
    exit(1)
    
ctx.calibrateADC()
ctx.calibrateDAC()
```

```python
ps = ctx.getPowerSupply()
ps.reset()
ps.enableChannel(0, True)
ps.pushChannel(0, 2.5)
```

```python
digital = ctx.getDigital()
buffer, sampling_params = StreamOutUtils.setAll(digital, bitstream, channels_out)
digital_sampling_freq = sampling_params.sampling_frequency
buffer_size = sampling_params.buffer_size
```

```python
digital.setDirection(dch_en_rd, libm2k.DIO_INPUT)
digital.enableChannel(dch_en_rd, False)
digital.setDirection(dch_clk_rd, libm2k.DIO_INPUT)
digital.enableChannel(dch_clk_rd, False)
digital.setDirection(dch_d_rd, libm2k.DIO_INPUT)
digital.enableChannel(dch_d_rd, False)
print(digital_sampling_freq)
digital.setSampleRateIn(digital_sampling_freq)
print(digital.getSampleRateIn())
```

```python
ain = ctx.getAnalogIn()
ain.reset()

ain.enableChannel(libm2k.CHANNEL_1, True)
ain.enableChannel(libm2k.CHANNEL_2, True)
ain.setSampleRate(MAX_SAMPLE_RATE)
ain.setOversamplingRatio(MAX_SAMPLE_RATE // digital_sampling_freq)

# configure trigger
atrig = ain.getTrigger()
atrig.reset()

atrig.setAnalogSource(0)
atrig.setAnalogMode(0, libm2k.ANALOG)
atrig.setAnalogLevel(0, 0.1)
atrig.setAnalogCondition(0, libm2k.RISING_EDGE_ANALOG)
atrig.setDigitalCondition(0, libm2k.SRC_NONE)
```

```python
trig = digital.getTrigger()
trig.reset()
trig.setDigitalDelay(0)
trig.setDigitalSource(libm2k.SRC_NONE)
trig.setDigitalCondition(dch_en_wr, libm2k.RISING_EDGE_DIGITAL)
```

```python
ctx.startMixedSignalAcquisition(buffer_size)
digital.push(buffer)
digital_data, analog_data = [], []
digital_data.extend(digital.getSamples(buffer_size))
analog_data.extend(ain.getSamples(buffer_size))
ctx.stopMixedSignalAcquisition()

libm2k.contextClose(ctx, True)
```

```python
digital_enable_read = list(
    map(lambda s: ((0x0001 << dch_en_rd) & int(s)) >> dch_en_rd, digital_data)
)
digital_clk_read = list(
    map(lambda s: ((0x0001 << dch_clk_rd) & int(s)) >> dch_clk_rd, digital_data)
)
digital_data_read = list(
    map(lambda s: ((0x0001 << dch_d_rd) & int(s)) >> dch_d_rd, digital_data)
)
```

```python
OFFSET = 1.5
time_step = (1 / digital_sampling_freq) * 1000  # in ms
stop_time = buffer_size * time_step
time_axis = np.arange(0, stop_time, time_step, dtype=float)

plt.plot(
    time_axis, np.array(digital_data_read) + 5.5, label="DATA-DRD", color="magenta"
)
plt.plot(
    time_axis, np.array(digital_clk_read) + 7, label="CLOCK-DRD", color="#1f77b4"
)
plt.plot(
    time_axis, np.array(digital_enable_read) + 8.5, label="ENABLE-DRD", color="red"
)

plt.plot(analog_data[1], label="DATA-ARD", color="blue")
plt.plot(analog_data[0] + 3, label="ENABLE-ARD", color="green")


plt.legend(loc="lower left", reverse=True)
plt.xlabel("Time (ms)")
plt.grid(True)
plt.tight_layout(h_pad=0.5)
plt.show()
```
